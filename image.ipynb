```python
# build a tensor network to classify images with pytorch
# 1. load data
# 2. build a network
# 3. train the network
# 4. test the network

import torch
import torchvision
import torchvision.transforms as transforms
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim

def ImportTinyImageSetFromWeb():
    # load the data from web
    # transform the data to tensor
    # normalize the data
    # split the data into trainset and testset
    # return the trainset and testset
    transform = transforms.Compose(
        [transforms.ToTensor(),
            transforms.Normalize((0.5,0.5,0.5),(0.5,0.5,0.5))]
    )
    trainset = torchvision.datasets.CIFAR10(root='./data',train=True,download=True,transform=transform)
    trainloader = torch.utils.data.DataLoader(trainset,batch_size=4,shuffle=True,num_workers=2)
    testset = torchvision.datasets.CIFAR10(root='./data',train=False,download=True,transform=transform)
    testloader = torch.utils.data.DataLoader(testset,batch_size=4,shuffle=False,num_workers=2)
    classes = ('plane','car','bird','cat','deer','dog','frog','horse','ship','truck')
    return trainloader,testloader,classes

class Net(nn.Module):
    # build a network
    def __init__(self):
        super(Net,self).__init__()
        # 3 input image channel, 6 output channels, 5x5 square convolution
        # kernel
        self.conv1 = nn.Conv2d(3,6,5)
        self.pool = nn.MaxPool2d(2,2)
        self.conv2 = nn.Conv2d(6,16,5)
        # an affine operation: y = Wx + b
        self.fc1 = nn.Linear(16*5*5,120) # 5*5 from image dimension
        self.fc2 = nn.Linear(120,84)
        self.fc3 = nn.Linear(84,10)
    
    def forward(self,x):
        # define the forward process
        x = self.pool(F.relu(self.conv1(x)))
        x = self.pool(F.relu(self.conv2(x)))
        # reshape the tensor
        x = x.view(-1,16*5*5)
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        # no activation function for the last layer
        x = self.fc3(x)
        return x
```
